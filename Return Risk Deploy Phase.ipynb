```python
import pickle
import pandas as pd

# Load model
model = pickle.load(open("return_pipeline.pkl", "rb"))

# Example input
data = pd.DataFrame([{
    "product_category": "Toys",
    "previous_returns_count": 1,
    "price": 150.5,
    "customer_tenure_days": 500,
    "customer_total_orders": 10,
    "order_hour": 14,
    "seller_rating": 4.5
}])

# Make prediction
pred = model.predict(data)[0]
proba = model.predict_proba(data)[:,1][0]

print("Prediction:", pred)
print("Probability:", proba)
```
